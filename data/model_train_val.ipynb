```python
# Validation metrics

from ultralytics import YOLO
from glob import glob
import yaml

conf = 0.25
iou = 0.6
experiment_path = '/home/revanth/Yolo/ultralytics/data/Tunnel/First4/'
data_path = '/home/revanth/Yolo/ultralytics/data/final_train_data/data.yaml'

with open(data_path, 'r') as file:
    data = yaml.safe_load(file)

val_images = glob(data['val']+'/*.png')

model = YOLO(f"{experiment_path}weights/best.pt")

metrics = model.val(data=data_path, imgsz=640, batch=32, conf=conf, iou=iou)
metrics.box.map
metrics.box.map50
metrics.box.map75
metrics.box.maps

for each_val_img in val_images:
    results = model.predict(each_val_img, conf=conf, iou=iou, save=True)
```
